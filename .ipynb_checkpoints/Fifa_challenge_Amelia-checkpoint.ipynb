```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
fifa_df = pd.read_csv('fifa21_train.csv')
pd.set_option("display.max_columns", None)
fifa_df.head(50)
```

```python
fifa_df.describe()
```

```python
print(fifa_df.shape)
print(list(fifa_df.columns))
```

```python
#Standarizing header names

fifa_clean = fifa_df.copy()
fifa_clean.columns = list(map(lambda x: x.lower(), fifa_clean.columns))
fifa_clean.columns = list(map(lambda x: x.replace(" ", "_"), fifa_clean.columns))
fifa_clean.columns = list(map(lambda x: x.replace("/", "_"), fifa_clean.columns))
fifa_clean.rename(columns = {'team_&_contract': 'team'}, inplace = True)
print(list(fifa_clean.columns))

#Checking duplicates on cols by sorting
cols_original = fifa_clean.columns
cols_ordered = list(fifa_clean.columns)
cols_ordered.sort()
fifa_clean = fifa_clean[cols_ordered]
print(list(fifa_clean.columns))
```

```python
# Going back to original cols order
fifa_clean = fifa_clean[cols_original]
print(list(fifa_clean.columns))

for v, i in enumerate(fifa_clean.columns):
    print(v,i)
```

```python
# Droping duplicates
print('Before:', fifa_clean.shape)
fifa_clean.drop_duplicates()
print('After:', fifa_clean.shape)
```

```python
# Checking data types
print('Numerical cols are:', list(fifa_clean._get_numeric_data().columns),'\n')
print('Categorical cols are:', list(fifa_clean.select_dtypes(['object']).columns))
```

```python
# Do we have nan values?
fifa_nan_categorical = fifa_clean.select_dtypes(['object']).isna().sum()
print('Nan categorical: \n', fifa_nan_categorical)
```

```python
fifa_nan_numerical = fifa_clean.select_dtypes(['int', 'float']).isna().sum()
print('Nan numerical: \n', fifa_nan_numerical)
```

```python
# Cleaning cols
cols_ame = fifa_clean.iloc[:, 34:67]
print(cols_ame.dtypes)
```

```python
# Handling special chars
cols_ame[['w_f','sm', 'ir']] = cols_ame[['w_f','sm', 'ir']].replace('[^\d]', '', regex=True)
display(cols_ame.head(50))
```

```python
nan = cols_ame.isna().sum().head(34)
nan_cols_name = nan.loc[nan > 0]
print(nan_cols_name)

#Categorical replace nan by most frequent value
most_freq_a_w = cols_ame['a_w'].value_counts().idxmax()
cols_ame['a_w'] = cols_ame['a_w'].fillna(most_freq_a_w)
most_freq_d_w = cols_ame['d_w'].value_counts().idxmax()
cols_ame['d_w'] = cols_ame['d_w'].fillna(most_freq_d_w)

nan = cols_ame.isna().sum().head(34)
```

```python
cols_ame[['vision', 'composure', 'balance', 'jumping', 'interceptions', 'positioning', 'sliding_tackle']].describe()
```

```python
# What are we going to do with nan values?
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize = (10,8), dpi = 80)
ax= sns.boxplot(data=cols_ame[['vision', 'composure', 'balance', 'jumping', 'interceptions', 'positioning', 'sliding_tackle']], 
                 orient="v", palette="Set2")

plt.show()

# Looks like we have extreme values. As this kind of data has an important impact over mean, we'll fill nan values with median
numerical_nan = ['vision', 'composure', 'balance', 'jumping', 'interceptions', 'positioning', 'sliding_tackle']

for i in numerical_nan:
    mean_val = np.mean(cols_ame[i])
    cols_ame[i] = cols_ame[i].fillna(mean_val)
    
```

```python
nan = cols_ame.isna().sum().head(34)
nan
```

```python
# CLUBS
print('Before:', fifa_clean['club'].isna().sum())
fifa_clean['club'] = fifa_clean['club'].fillna('Unknown')
print('After:', fifa_clean['club'].isna().sum())

clubs_list = fifa_clean['club'].to_list()
countries_list = fifa_clean['nationality'].to_list()
print(len(clubs_list))
print(len(countries_list))

for i in range(len(countries_list)):
    if clubs_list[i] == countries_list[i]:
        clubs_list[i] = "Unkown"

fifa_clean['club'] = clubs_list
```

```python
print('Before', fifa_clean['height'].unique())
print('Before', fifa_clean['weight'].unique())

fifa_clean['weight'] = fifa_clean['weight'].replace('lbs', '', regex=True)
fifa_clean['height'] = fifa_clean['height'].str.replace('[^\d]', '')
fifa_clean['height'] = fifa_clean['height'].str.rsplit(", ", 1)
print('After', fifa_clean['weight'].unique())
print('After', fifa_clean['height'].unique())


```
